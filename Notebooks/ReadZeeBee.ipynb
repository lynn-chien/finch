```python
# Loop through all the row in the PD 

currentWAV = []
isound = 0
normalize = True
plotMe = False
MIN_LEN = 425      # Min length of segment in sampling points - 
MIN_PAD = 425     # Min length needed for spectrogram resolution in sampling rate- padded if less

for irow, row in soundsepPD.iterrows():
    # Check if h5 file exists
    fileRoot = (row['file.name'].split('/')[-1]).split('.')[0]
    filename = fileRoot + '_' + str(row['project.start_index'])
    fh5name = 'h5filesSpectro/%s.h5' % (filename)
    
    if os.path.exists(fh5name):
        continue

    # read the wav file if needed
    if (row['file.name'] != currentWAV):
        currentWAV = row['file.name']
        currentWAVLocal = '.' + currentWAV.split('/Feb 22, 2019')[1]
        soundIN = WavFile(file_name=currentWAVLocal)      
        
    if (soundIN.num_channels > 1):
        print('WAV file has more than 1 channel')
        break
        
    dt = row['file.stop_index']- row['file.start_index']
    if (dt < MIN_LEN) :
        print('%s Too short' % fh5name)
        continue
    elif (dt < MIN_PAD):
        ptadd = int((MIN_PAD - dt)/2 + 1)
        
        istart = row['file.start_index']-ptadd
        if istart < 0 :
            print ('Skipping short syllable at start of file')
            continue
        iend = row['file.stop_index'] + ptadd
        if (iend > len(soundIN.data) ):
            print ('Skipping short syllable at eend of file')
            continue
    else:
        istart = row['file.start_index']
        iend = row['file.stop_index']
            
    soundSYL = soundIN.data[istart:iend]
    isound += 1
        
    fileRoot = (row['file.name'].split('/')[-1]).split('.')[0]
    filename = fileRoot + '_' + str(row['project.start_index'])
    
    birdname = row['source.name']
    calltype = 'So'
        
        
    # Normalize if wanted
    rms = soundIN.data.std() 
    if normalize :
        maxAmp = rms
    else:
        maxAmp = 1.0
    
    # Create BioSound Object
    myBioSound = BioSound(soundWave=soundSYL.astype(float)/maxAmp, fs=float(soundIN.sample_rate), emitter=birdname, calltype = calltype)
             
    # Calculate the spectrogram and the rms
    myBioSound.spectroCalc(spec_sample_rate=1000, freq_spacing = 50, min_freq=0, max_freq=10000)
    myBioSound.rms = rms  # The unnormalized intensity
    if (myBioSound.spectro.shape[0] != 196):
        print('Problem with frequency resolution for', filename)
        print('Spectrogram shape is', myBioSound.spectro.shape)
        break
       
    # Calculate amplitude enveloppe
    myBioSound.ampenv(cutoff_freq = 50, amp_sample_rate = 1000)
       
    # Calculate the power spectrum
    myBioSound.spectrum(f_high=10000)
    
    # Calculate fundamental and related values.  These are the default values.
    # For the estimation of the fundamental, four methods are available: 
    # 'AC' - Peak of the auto-correlation function
    # 'ACA' - Peak of envelope of auto-correlation function 
    # 'Cep' - First peak in cepstrum 
    # 'Stack' - Fitting of harmonic stacks (default - works well for zebra finches)
    
    myBioSound.fundest(maxFund = 1500, minFund = 300, lowFc = 200, highFc = 6000, 
                           minSaliency = 0.5, debugFig = 0, 
                           minFormantFreq = 500, maxFormantBW = 500, windowFormant = 0.1,
                           method='Stack')
      
    # Calculate the MPS
    myBioSound.mpsCalc(window=0.01, Norm = True)
                
    if plotMe and isound <= 10: 
        print('++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
        print('                Bird %s    Call Type %s' % (myBioSound.emitter, myBioSound.type))
        myBioSound.plot(DBNOISE=50, f_low=250, f_high=10000)  
        print('++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')

    # Save the results
    fh5name = 'h5filesAmp/%s.h5' % (filename)
    myBioSound.saveh5(fh5name)  
    if ((isound % 1000) == 0):
        print('Done w', isound, '/', nsounds )

```
